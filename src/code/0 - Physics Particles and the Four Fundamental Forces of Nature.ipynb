<b>- Pauli Exclusion Principle</b>

This Principle tells that all particles can be divided into two types, `Bosons` and `Fermions`, accordingly to their behavior in the space:

> **Bosons** - can occupy the same space, that is, the `superposition` is true between them and do not follow the `Pauli Exclusion`;

> **Fermions** - cannot occupy the same space, that is, the `superposition` is not true between them and do not follow the `Pauli Exclusion`.

<br><br>

<b>- Properties</b>

The Fundamental Particles have a bunch of properties that describe their uniqueness, being the `Spin`, `Color`, `Flavor`, `Charge` and `Mass` the main ones. Let's see a briefly explanation for each one of them:

> **Spin** - it is not like the rotation spin you are picturing right now, but you can take this thought in consideration to understand what spin is. `Fermions` will always have non-integer spins, such as `1/2, 3/2 and 5/2`, while `Bosons` will always have integer spins, such as `0, 1 and 2`.;

>> Fermions gotta take two full rotations to get the same `Wave Function` value to the origin value in the beggining of the spin (*i.e. 1, 2, 3, 2, 1*). However, Bosons gotta take only a full rotation to get the same `Wave Function` value to the origin one (*i.e. 1, 2, 1*).

> **Color** - one more time, it is not like the visual colors you are thinking, but you can take this thought in consideration (*yeah, I know, quantum is confusing*). Only `quarks` and `gluons` have colors;

> **Flavor** - no, you cannot taste the Fundamental Particles - at least so far üòÇ. Flavor is a property present only for `quarks` and `leptons` that tells if the particle is an `electron`, `tau` or `muon` and `electron neutrino`, `tau neutrino` or `muon neutrino`;

> **Charge** -  finally, one property we are already used to. Charge is the electrical energy of them, being positive or negative. All `quarks`, `non-neutrino leptons` and `W bosons` have charge;

> **Mass** - probably this is the most known property of these particles. Mass is literally the mass of the particles. All `quarks`, `leptons`, `Z bosons`, `W bosons` and `Higgs bosons` have mass.

<h2 id='fermions' style='color:#7159c1'>2) Fermions</h2>

Now let's take a deep understanding about the Fermions: `quarks`, `leptons` and `baryons`.

<br>

<b>- Quarks</b>

- Quarks are the elementary particles that give origin to the `baryons`, such as `protons` (*two quarks up and one down*) and `neutrons` (*one quark up and two downs*);

- `Lambdas, sigmas, xis and omegas` are other baryons made up of quarks;

- They come in six different types or "flavors": `up, down, charm, strange, top, and bottom`;

- Carry fractional electric charges, like `+2/3` or `-1/3`;

- They are always found in combinations of two or three to form larger particles.

<br>

<b>- Leptons</b>

- Leptons are also elementary particles and another type of fundamental matter particle;

- They come in three generations: the `electron and electron neutrino`, the `muon and muon neutrino`, and the `tau and tau neutrino`;

- `Muon and muon neutrino` and `tau and tau neutrino` are instables, are created while colliding particles and their life-time is significantly short;

- `Muons` and `taus` can decay to `electrons`, while `muon neutrinos` and `tau neutrinos` can decay to `electron neutrinos` via `Weak Force`;

- Leptons come in pairs of electron/muon/tau and its respective neutrino;

- Leptons have integral electric charges: `-1 for electrons`, `-2 for muons`, and `-3 for taus`;

- Electrons are the most familiar leptons and are involved in electricity, chemical reactions and the formation of atoms.

<br>

<b>- Baryons</b>

- Baryons are a type of particle made up of `three quarks`;

- `Protons and neutrons` are the most common and stable baryons found in atomic nuclei;

- Others known baryons are `sigma, lambda, xi and omega`. They are instables, are created while colliding particles and their life-time is significantly short;

- Baryons play a crucial role in the structure of matter and contribute to the diversity of atomic nuclei.


```
\ Proton: It is composed of two up quarks and one down quark (uud) - +1 electric charge.

\ Neutron: It is composed of one up quark and two down quarks (udd) - 0 electric charge.

\ Œ£+ (Sigma-plus): It is made up of two up quarks and one strange quark (uus) - +1 electric charge.
\ Œ£0 (Sigma-zero): It consists of one up quark, one down quark, and one strange quark (uds) - 0 electric charge.
\ Œ£- (Sigma-minus): It is composed of one down quark and two strange quarks (dss) - -1 electric charge.

\ Œû0 (Xi-zero): It consists of one up quark, one down quark, and one strange quark (uds) - 0 electric charge.
\ Œû- (Xi-minus): It is composed of one down quark and two strange quarks (dss) - -1 electric charge.

\ Œ©- (Omega-minus): It is made up of three strange quarks (sss) - -1 electric charge.

\ Œõ (Lambda): It consists of one up quark, one down quark, and one strange quark (uds) - 0 electric charge.
```

<h2 id='bosons' style='color:#7159c1'>3 - Bosons</h2>

It is the Bosons turn and we will see info about the `gauge/scalar bosons` and the `mesons`.

<br>

<b>- Gauge/Scalar Bosons</b>

- Bosons are particles that mediate the `Fundamental Forces of Nature` - Strong, Weak, Electromagnetic and Gravitational Forces;

- The `photon` is a boson that carries the `Electromagnetic Force` - `Gauge Boson`;

- The `W and Z bosons` are involved in the `Weak Nuclear Force` - `Gauge Boson`;

- The `gluon` is the boson that carries the `Strong Nuclear Force` - `Gauge Boson`;

- The `Higgs boson` is responsible for giving other particles `Mass` - `Scalar Boson`;

- The `gravitron` is a hypothetical boson that carries the `Gravitational Force` - hoping that this one be confirmed soon, by the way, when this particle be confirmed, it will be a `Gauge Boson`!!

- `Gauge Bosons` are all bosons that carry a Fundameental Force of Natura. `Scalar Bosons` are all bosons that does not carry a Fundamental Force and have a spin equals to `zero`.

<br>

<b>- Mesons</b>

- Mesons are type of boson particles that are made up of a `quark` and its respective `anti-quark`;

- The most known meson is the `pion`;

- Pions carry the `Residual Strong Force`, making the protons and neutrons be together and forming the atom nuclei.

<h2 id='fundamenta-forces-of-nature' style='color:#7159c1'>4 - Fundamental Forces of Nature</h2>

The Four Fundamental Forces of Nature are composed by the `Strong Force`, the `Weak Nuclear Force`, the `Electromagnetic Force` and the `Gravitational Force`. Let's see each one of them!

<br>

<b>- Strong Force (`changes color property`)</b>

- It is mediated/carried by the `gluons`;

- Since gluons have the `color` property, they are responsible to change the colors of quarks when they are changed between the quarks;

- Also, gluons are responsible to bind quarks together and, consequently, form `baryons`;

- The Strong Force can also act in a form of `Residual Strong Force`;

- The `Residual Strong Force` is when a quark tries to get out of a baryon, such as proton and neutron, but the gluons does not allow that, making only a fraction of energy get away. This energy is converted into a `pion meson` that carries gluons to bind the quark and its anti-quark together into the pion. When this pion reaches another baryon, the anti-quark destroys itself with its respective quark present in this another baryon and the remaining pion quark binds to the baryon occuping the space of the destroyed quark - RIP destroyed quark üò¢.

<br>

<b>- Weak Nuclear Force (`changes flavor property`)</b>

- It is mediated/carried by the `Z, W- and W+`;

- These bosons are responsible to the particles decayment, such as `muon and taus decaying to electrons`, `muon neutrinos and tau neutrinos decaying to electron neutrinos`, `charm and top quarks decaying to up quark` and `strange and bottom quarks decaying to down quark`;

- Also, they are responsible for some `Beta Decays`, when a proton decays to a neutron (*an up quark decays to a bottom quark*) and when a neutron decays to a proton (*a bottom quark decays to an up quark*);

- Some examples of `Z Boson` decayment pairs are: `electron and positron (anti-electron)`, `muon and anti-muon`, `tau and anti-tau`, `electron/muon/tau neutrino and anti-electron/muon/tau neutrino` and `quark and anti-quark`;

- About the `W- and W+ bosons`, neutrinos have both of them inside of its body. When a Beta Decayment happens, the `W+` gets out of the neutrino that is transformed into an electron. This W+ boson reaches a neutron, gets into a `down quark` transforming it into an up quark and, consequently, decaying the neutron into a proton - electric charges mutations: -1/3 (bottom quark) + 1 (W+ boson) = 2/3 (up quark);

- The inverse is also true, that is, the `W-` gets out of the neutrino that is transformed into a positron. This W- boson reaches a proton, gets into an `up quark` transforming it into a bottom quark and, consequently, decaying the proton into a neutron - electric charges mutations: +2/3 (up quark) - 1 (W-boson) = -1/3 (bottom quark).

<br>

<b>- Electromagnetic Force (`changes charge property`)</b>

- It is mediated/carried by the `photon`;

- These bosons are responsible for carrying `heat`, `light`, `magnetism` and `electric charge` between the particles;

- Also, can act as `particles` and `waves`;

- The carried charges can be one of the two types: `positive (+)` and `negative (-)`;

- Usually, different charges attract themselves, whereas similar charges repel themselves.

```
\ + and +: REPEL
\ + and -: ATTRACT
\ - and -: ATTRACT
\ - and +: REPEL
```

<b>- Gravitational Force (`changes space and time. interacts with the mass property`)</b>

- The `Gravitational Force` is the unique force that does not have a Quantum Explanation yet and, to learn about it, we have to loo at the `Relativity Theory` of Albert Einstein;

- It is believed that this force is mediated/carried by the `gravitrons`, hypothetically bosons that have not been obseerved and and have not gotten the existence confirmed by Scientists yet;

- This forces changes the `space and time dimensions`, shortening the space and slowing down the time;

- Also, this force can only be applied to particles that have mass;

- All massive particles get their mass via `Higgs Bosons`, the unique `Scalar Bosons` discovered so far!

<h2 id='bonus-antimatter-and-dark-matter' style='color:#7159c1'>5 - Bonus: Antimatter and Dark Matter</h2>

In this bonus section we will talk about two types of matters that intrigue Scientists: `Antimatter` and `Dark Matters`.

<br>

<b>- Antimatter</b>

- `Antimatter` is rare in the observable universe compared to ordinary matter;

- All `fermions` have their respective antimatter, however, `bosons` does not have;

- Small amounts of antimatter have been produced and detected in laboratory settings, such as `positrons` (*antielectrons*) and `antiprotons`;

- In natural cosmic conditions, antimatter is thought to be produced through `high-energy processes`, such as `particle collisions` and `certain types of radioactive decays`;

- However, the abundance of naturally occurring antimatter in the universe is still an open question, as it has not been observed in large quantities outside of laboratory-created environments.

<br>

<b>- Dark Matter</b>

- `Dark Matter` is thought to be present throughout the universe on large scales, including galaxies, galaxy clusters, and the cosmic web;

- It does not emit, absorb, or interact significantly with `Electromagnetic Radiation`, making it challenging to detect directly;

- Also, these kind of matter are considered `cold`, the inverse of the Standard Particles seen so far, that are considered `hot`. Cold particles are those ones that travel in `low speed`, that is, `slow compared to the light`. Hot particles are those ones that travel in `high sped`, that is, `almost to the speed of light`;

- Observations of `galaxy rotation curves`, `gravitational lensing`, and the `cosmic microwave background` provide evidence for the presence of dark matter;

- Dark matter is believed to play a crucial role in the formation and evolution of galaxies and the large-scale structure of the universe.

````python
#---- Particles DF ----#
features = [
    'pdg_id', 'pdg_name', 'name', 'charge'
    , 'rank',  'quarks'
    , 'mass', 'mass_lower', 'mass_upper'
    , 'width', 'width_lower', 'width_upper'
]

particles_names = [
    #---- Quarks ----#
    'down', 'up', 'strange', 'charm', 'bottom', 'top'
    
    #---- Anti-Quarks ----#
    , 'anti_down', 'anti_up', 'anti_strange', 'anti_charm', 'anti_bottom', 'anti_top'
    
    #---- Leptons ----#
    , 'electron', 'electron_neutrino', 'muon', 'muon_neutrino', 'tau', 'tau_netrino'
    
    #---- Anti-Leptons ----#
    , 'anti_electron (positron)', 'anti_electron_neutrino', 'anti_muon'
    , 'anti_muon_neutrino', 'anti_tau', 'anti_tau_neutrino'
    
    #---- Baryons ----#
    , 'proton', 'neutron'
    
    #---- Anti-Baryons ----#
    , 'anti_proton', 'anti_neutron'
    
    #---- Pions ----#
    , 'pion+', 'pion0', 'pion-'
    
    #---- Gauge Bosons ----#
    , 'photon', 'W+', 'W-', 'Z0', 'gluon'
    
    #---- Scalar Bosons ----#
    , 'higgs'
    
    #---- Other Particles ----#
    , 'kaon+', 'kaon-', 'lambda0', 'sigma+', 'sigma0', 'sigma-'
    , 'xi0', 'xi-', 'omega-'
]

particles_df = pd.DataFrame(columns=features)
particles_df
````

````python
#---- Adding Datas ----#
for key, particle_name in zip(particles_dictionary.keys(), particles_names):
    temp_data = pd.DataFrame(data=[[
         particles_dictionary[key].pdgid
         , particles_dictionary[key].pdg_name
         , particle_name
         , particles_dictionary[key].charge

         , particles_dictionary[key].rank
         , particles_dictionary[key].quarks


         , particles_dictionary[key].mass
         , particles_dictionary[key].mass_lower
         , particles_dictionary[key].mass_upper

         , particles_dictionary[key].width
         , particles_dictionary[key].width_lower
         , particles_dictionary[key].width_upper
    ]], columns=features)
    
    particles_df = pd.concat([particles_df, temp_data], ignore_index=True)
    
particles_df
````

````python
#---- DataFrame Index ----#
#
# - converting 'pdg_id' feature into the index
# - sorting the index into ascending order (anti-matters to matters)
#
particles_df.set_index('pdg_id', inplace=True)
particles_df.sort_index(inplace=True)
particles_df
````

````python
#---- Replacing Missing and Empty Values ----#
#
# - replacing both of them by hyphen '-'
#
particles_df.fillna('-', inplace=True)
particles_df.replace('', '-', inplace=True)
````
